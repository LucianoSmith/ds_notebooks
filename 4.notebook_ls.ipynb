```python
# Investigación Operativa 3 - FIUBA - 13/OCT/2022
# Demo Local Search -  https://github.com/Akavall/AntColonyOptimization -

import numpy as np
from ant_colony import AntColonyLS as ACLS
import matplotlib.pyplot as plt # funciones para graficar (presentación gráfica)
from matplotlib.collections import LineCollection # funciones para graficar recorrido (presentación gráfica)


# carga de datos de las ciudades

pos = [[float(x) for x in s.split()[1:]] for s in open('data/tsp76.txt').readlines()]       # pos: posiciones de las ciudades las carga desde el archivo
n = len(pos)                                                                                # n: cantidad de ciudades

# calcula la matriz de distancias
adj_mat = np.ones((n,n)) * np.inf                                                           # genera una matriz adj_mat llena de ceros
for i in range(n):
    for j in range(i, n):
        if (i!=j):
            adj_mat[i][j] = adj_mat[j][i] = np.linalg.norm(np.subtract(pos[i], pos[j]))     # calcula la distancia euclidiana
```

```python
s = ([(0, 32), (32, 57),(57, 24),(24, 45),(45, 37),(37, 52),(52, 71),(71, 13),(13, 61),(61, 5),(5, 66),(66, 33),(33, 74),(74, 16),(16, 42),(42, 41),(41, 73),(73, 44),(44, 12),(12, 53),
  (53, 15),(15, 2),(2, 55),(55, 17),(17, 4),(4, 47),(47, 28),(28, 25),(25, 59),(59, 36),(36, 20),(20, 6),(6, 75),(75, 11),(11, 19),(19, 26),(26, 34),(34, 68),(68, 14),(14, 70),(70, 63),
  (63, 43),(43, 31),(31, 50),(50, 23),(23, 64),(64, 30),(30, 38),(38, 8),(8, 9),(9, 7),(7, 56),(56, 69),(69, 72),(72, 10),(10, 21),(21, 48),(48, 39),(39, 60),(60, 27),(27, 29),(29, 1),
  (1, 22),(22, 54),(54, 18),(18, 58),(58, 51),(51, 3),(3, 67),(67, 65),(65, 62),(62, 40),(40, 49),(49, 46),(46, 35),(35, 0)], 1925.846464390129)

# Imprime el recorrido

fig, ax = plt.subplots()
ax.set_xlim(0, 80)
ax.set_ylim(0, 80)

lines = [[pos[int(s[0][_][0])], pos[int(s[0][_][1])]] for _ in range(n)]
line_segments = LineCollection(lines, color='r')
ax.add_collection(line_segments)
```

```python
ls = ACLS(adj_mat, n, 1, 100, 0.95, 1, 1)        
# instancio la clase (nada que ver con ACO)
```

```python
# Aplico local search 
sp = ls.local_search(s)
```

```python
# Imprime el recorrido encontrado

fig, ax = plt.subplots()
ax.set_xlim(0, 80)
ax.set_ylim(0, 80)

lines = [[pos[int(sp[0][_][0])], pos[int(sp[0][_][1])]] for _ in range(n)]
line_segments = LineCollection(lines, color='b')
ax.add_collection(line_segments)
```
