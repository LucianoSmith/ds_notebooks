```python
import random
import os
import shutil

```

```python


folder1 = "./dataset_vgg1/train/h/"
folder2 = "./dataset_vgg1/valid/h/"
for i in range(347):
    cat_image_name = random.choice(os.listdir(folder1))
    # mueve el archivo a     
    print(i, "] mover ", folder1 + cat_image_name, " a ", folder2 + cat_image_name)
    shutil.move(folder1 + cat_image_name, folder2 + cat_image_name)  
```
